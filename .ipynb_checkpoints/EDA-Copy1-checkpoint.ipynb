```python
from __future__ import unicode_literals
import os
import codecs
import pandas as pd
import numpy as np
from email.parser import Parser

%matplotlib inline
from matplotlib import pyplot as plt

import seaborn as sns
sns.set(color_codes=True)

import regex as re
```

## I - Data Extraction

### Extract and store all files into pandas dataframe

```python
# fpath = '20_newsgroups/alt.athesim_comp'
fpath = '20_newsgroups'
messagefolder =[]

for root, dirs, files in os.walk(fpath, topdown=False):
    for name in files:
        filename = os.path.join(root, name)
        messagecat = os.path.basename(os.path.dirname(filename))

        with codecs.open(filename, 'r', encoding='utf_8', errors='ignore') as f:
            messagefolder.append({"Category":messagecat, "Body":f.read()})
        
            
messagefolder = pd.DataFrame(messagefolder)
print("Number of documents: {}".format(len(messagefolder)))
print(messagefolder.tail())
```

### Extract features and body of each email 

```python
# headers = Parser().parse(open(messagefile, 'r'))
def message_parser(messagefile, cat):
    headers = Parser().parsestr(messagefile,headersonly=True)

    # Get email's body
    if headers.is_multipart():
        for part in headers.get_payload():
            txt = part.get_payload()
    else:
        txt = headers.get_payload()

    # Get email's headers
    df = pd.DataFrame({"Category": [cat],
            'Path':[headers['Path']],
                      'From': [headers['From']],
                      'Newsgroups': [headers['Newsgroups']],
                      'Subject': [headers['Subject']],
                      'Date': [headers['Date']],
                      'Organization':[headers['Organization']],
                      'NNTP-Posting-Host':[headers['NNTP-Posting-Host']],
                      'References': [headers['References']],
                      'Message-ID': [headers['Message-ID']],
                      'Sender': [headers['Sender']],
                      'Body': [txt.replace('\n', '').strip()],
                      'Lines': [headers['Lines']]}, index=[0])
#     df['Lines'] = pd.to_numeric(df['Lines'])
#     print(type(headers['Lines']))
    return df
```

```python
frames = []
for idx,row in messagefolder.iterrows():
    frames.append(message_parser(row['Body'], row['Category']))
    
message_df = pd.concat(frames)
message_df = message_df.reset_index(drop=True)
print('Length of df: {}'.format(len(message_df)))
print(message_df.head())
```

#### Lemmatize, remove punctuations and stopwords

```python
def clean_text(msg):
    tokens = [token.lemma_ for token in msg]
    tokens = [tok for tok in tokens if (tok not in stopwords and tok not in punc)] 
    return tokens
```

```python
cat_df['token_lemma'] = cat_df['Body'].apply(nlp).apply(clean_text)
cat_df['unique_tokens']= cat_df['token_lemma'].apply(set).apply(len)
print(cat_df.head())
```

### Top 5 most frequent words per category

NOUN could be a good predictor of an email's category
- Implement Part-of-Speech tagging
- Extract NOUN words
- Obtain frequency of NOUN words per category
- Chart top 5 most frequent words per category

```python
import collections
```

```python
def tag_text(text):
    text = nlp(text)
    tokens = [str(token) for token in text if token.pos_ == "NOUN"]
    tokens = [tok for tok in tokens if (tok not in stopwords and tok not in punc)] 
    return tokens
```
